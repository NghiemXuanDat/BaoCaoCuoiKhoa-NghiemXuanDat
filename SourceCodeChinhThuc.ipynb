```python
from tkinter import *
from tkinter import messagebox
import tkinter as tk
from tkinter import ttk

# Tạo ra cửa sổ ứng dụng
window = Tk()

# Set up thông tin cho window 
window.title("Hệ thống quản lý thông tin học sinh")
window.geometry("800x600")

list_hocsinh = []

# Tạo label (3)
name_label = Label(window, text="Tên học sinh:") 
mahocsinh_label = Label(window, text="Mã học sinh:")
phuhuynh_label = Label(window, text="Tên phụ huynh:")

# Tạo Entry (6)
name_entry = Entry(window)
mahocsinh_entry = Entry(window)
phuhuynh_entry = Entry(window)
search_name_entry = Entry(window)
search_mahocsinh_entry = Entry(window)
search_phuhuynh_entry = Entry(window)

# Dán các widget lên window
name_label.grid(row=0, column=0)
mahocsinh_label.grid(row=1, column=0)
phuhuynh_label.grid(row=2, column=0)
name_entry.grid(row=0, column=1)
mahocsinh_entry.grid(row=1, column=1)
phuhuynh_entry.grid(row=2, column=1)
search_name_entry.grid(row=4, column=1)
search_mahocsinh_entry.grid(row=4, column=3)
search_phuhuynh_entry.grid(row=4, column=5)


# Khai báo class và hàm init
class Student:
    def __init__(self, hocsinh, mahocsinh, phuhuynh):
        self.hocsinh = hocsinh
        self.mahocsinh = mahocsinh
        self.phuhuynh = phuhuynh

    

    def them_hocsinh():
        hocsinh_value = name_entry.get()
        mahocsinh_value = int(mahocsinh_entry.get())
        phuhuynh_value = phuhuynh_entry.get()
        def check_ma_hocsinh(ma_hs):
            for student in list_hocsinh:
                if student.mahocsinh == ma_hs:
                    return True
            return False
        
        if check_ma_hocsinh(mahocsinh_value):
            messagebox.showerror("Error", "Mã học sinh đã tồn tại")
        else:
            new_student = Student(hocsinh_value, mahocsinh_value, phuhuynh_value)
            list_hocsinh.append(new_student)
            print([item.hocsinh for item in list_hocsinh])
            print([item.mahocsinh for item in list_hocsinh])
            print([item.phuhuynh for item in list_hocsinh])
            messagebox.showinfo("Thêm Học Sinh", f"{hocsinh_value}, {mahocsinh_value} và {phuhuynh_value} được thêm thành công")
            
            name_entry.delete(0, END)
            mahocsinh_entry.delete(0, END)
            phuhuynh_entry.delete(0, END)

    
    def search_name():
        name_search = search_name_entry.get()
        is_exist = False
        for student in list_hocsinh:
            if student.hocsinh == name_search:
                is_exist = True
                messagebox.showinfo("Tìm tên học sinh", f"Học sinh: {student.hocsinh}\nMã học sinh: {student.mahocsinh}\nPhụ huynh: {student.phuhuynh}")
                break

        if is_exist == False:
            messagebox.showinfo("Tìm tên học sinh", f"{name_search} không tìm thấy")

    def search_mahocsinh():
        mahocsinh_search = int(search_mahocsinh_entry.get())
        is_exist = False
        for student in list_hocsinh:
            if student.mahocsinh == mahocsinh_search:
                is_exist = True
                messagebox.showinfo("Tìm mã học sinh", f"Học sinh: {student.hocsinh}\nMã học sinh: {student.mahocsinh}\nPhụ huynh: {student.phuhuynh}")
                break

        if is_exist == False:
            messagebox.showinfo("Tìm mã học sinh", f"{mahocsinh_search} không tìm thấy")

    def search_phuhuynh():
        phuhuynh_search = search_phuhuynh_entry.get()
        is_exist = False
        for student in list_hocsinh:
            if student.phuhuynh == phuhuynh_search:
                is_exist = True
                messagebox.showinfo("Tìm phụ huynh", f"Học sinh: {student.hocsinh}\nMã học sinh: {student.mahocsinh}\nPhụ huynh: {student.phuhuynh}")
                break

        if is_exist == False:
            messagebox.showinfo("Tìm phụ huynh", f"{phuhuynh_search} không tìm thấy")

    def update_hocsinh():
        hocsinh_value = name_entry.get()
        mahocsinh_value = int(mahocsinh_entry.get())
        phuhuynh_value = phuhuynh_entry.get()
        is_exist = False
        for student in list_hocsinh:
            # Cập nhật thông tin phụ huynh
            if student.hocsinh == hocsinh_value and student.mahocsinh == mahocsinh_value:
                is_exist = True
                student.phuhuynh = phuhuynh_value 
                messagebox.showinfo("Update Student", f"Thông tin phụ huynh của {student.hocsinh} đã được cập nhật")
                break
            # Cập nhật thông tin học sinh
            if student.phuhuynh == phuhuynh_value and student.mahocsinh == mahocsinh_value:
                is_exist = True
                student.hocsinh = hocsinh_value
                messagebox.showinfo("Update Student", f"Thông tin học sinh {student.hocsinh} đã được cập nhật")
                break
            # Cập nhật thông tin mã học sinh
            if student.hocsinh == hocsinh_value and student.phuhuynh == phuhuynh_value:
                is_exist = True
                student.mahocsinh = mahocsinh_value 
                messagebox.showinfo("Update Student", f"Thông tin mã học sinh {student.hocsinh} đã được cập nhật")
                break
            # Cập nhật thông tin phụ huynh và học sinh
            if student.mahocsinh == mahocsinh_value:
                is_exist = True
                student.phuhuynh = phuhuynh_value 
                student.hocsinh = hocsinh_value 
                messagebox.showinfo("Update Student", f"Thông tin đã được cập nhật")
                break
            # Cập nhật thông tin học sinh và mã học sinh
            if student.phuhuynh == phuhuynh_value:
                is_exist = True
                student.hocsinh = hocsinh_value
                student.mahocsinh = mahocsinh_value
                messagebox.showinfo("Update Student", f"Thông tin đã được cập nhật")
                break
            # Cập nhật thông tin mã học sinh và phụ huynh
            if student.hocsinh == hocsinh_value:
                is_exist = True
                student.mahocsinh = mahocsinh_value 
                student.phuhuynh = phuhuynh_value
                messagebox.showinfo("Update Student", f"Thông tin đã được cập nhật")
                break
        if is_exist == False:
            messagebox.showinfo("Update Student", f"Thông tin này không tồn tại")

    def delete_hocsinh():
        hocsinh_value = name_entry.get()
        is_exist = False
        for student in list_hocsinh:
            if student.hocsinh == hocsinh_value:
                is_exist = True
                list_hocsinh.remove(student) 
                messagebox.showinfo("Xóa học sinh", f"{student.hocsinh} và thông tin liên quan đã được xóa")
                break

        if is_exist == False:
            messagebox.showinfo("Delete Student", f"{hocsinh_value} không tìm thấy")

    def reset():
        name_entry.delete(0, END)
        mahocsinh_entry.delete(0, END)
        phuhuynh_entry.delete(0, END)
        search_mahocsinh_entry.delete(0, END)
        search_phuhuynh_entry.delete(0, END)
        search_name_entry.delete(0, END)

    def update_treeview():
        # Clear the treeview
        for item in treeview.get_children():    
            treeview.delete(item)
        # Re-populate the treeview
        for student in list_hocsinh:
            treeview.insert("", tk.END, values=(student.hocsinh, student.mahocsinh, student.phuhuynh))        

    # Tạo button (8)
    them_hocsinh_bt = Button(window, text="Thêm học sinh", command=them_hocsinh)
    update_hocsinh_bt = Button(window, text="Cập nhật thông tin học sinh", command=update_hocsinh)
    delete_hocsinh_bt = Button(window, text="Xóa thông tin học sinh", command=delete_hocsinh)
    reset_list_bt = Button(window, text="Xóa thông tin thanh entry", command=reset)
    hienthi_treeview_bt = Button(window, text="Hiển thị treeview", command=update_treeview)
    search_name_bt = Button(window, text="Tìm tên học sinh", command=search_name)
    search_mahocsinh_bt = Button(window, text="Tìm mã học sinh", command=search_mahocsinh)
    search_phuhuynh_bt = Button(window, text="Tìm tên phụ huynh", command=search_phuhuynh)
    them_hocsinh_bt.grid(row=3, column=0)
    update_hocsinh_bt.grid(row=3, column=1)
    delete_hocsinh_bt.grid(row=3, column=2)
    hienthi_treeview_bt.grid(row=3, column=3)
    reset_list_bt.grid(row=3, column=4)
    search_name_bt.grid(row=4, column=0)
    search_mahocsinh_bt.grid(row=4, column=2)
    search_phuhuynh_bt.grid(row=4, column=4)

root = tk.Tk()

# Create Treeview widget
treeview = ttk.Treeview(root)
treeview['columns'] = ('hocsinh', 'mahocsinh', 'phuhuynh')

# Define column headings
treeview.column("#0", width=0, stretch=tk.NO)
treeview.column("hocsinh", anchor=tk.W, width=120)
treeview.column("mahocsinh", anchor=tk.CENTER, width=80)
treeview.column("phuhuynh", anchor=tk.W, width=120)

# Define column titles
treeview.heading("#0", text="", anchor=tk.W)
treeview.heading("hocsinh", text="Học Sinh", anchor=tk.W)
treeview.heading("mahocsinh", text="Mã Học Sinh", anchor=tk.CENTER)
treeview.heading("phuhuynh", text="Phụ Huynh", anchor=tk.W)

# Insert data into treeview
for student in list_hocsinh:
    treeview.insert("", tk.END, values=(student.hocsinh, student.mahocsinh, student.phuhuynh))

# Pack the treeview finally
treeview.pack(side=tk.TOP, fill=tk.X) 

    # Chạy ứng dụng
window.mainloop()
root.mainloop()
```
